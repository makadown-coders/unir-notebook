Inicio de codigo

```python
import cv2 as cv
img01=cv.imread('CochesVarios/Cars3.png')
```

Leer imagen en niveles de gris

```python
img02=cv.imread('CochesVarios/Cars3.png', cv.IMREAD_GRAYSCALE)
```

Procesando


```python

alto1, ancho1, canales1 = img01.shape
alto2, ancho2 = img02.shape

```

imprimiendo y mostrando imagen

```python
print("alto: ", alto1, " ancho: ", ancho1, " canales: ", canales1)
print("alto: ", alto2, " ancho: ", ancho2)

cv.imshow("img01", img01)
cv.imshow("img02", img02)
```

cerrando imagenes

```python
cv.waitKey(0)
cv.destroyAllWindows()
```

mostrando imagen con matplotlib

```python
from matplotlib import pyplot as plt
plt.imshow(cv.cvtColor(img01, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```
