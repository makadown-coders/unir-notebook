```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
img01=cv.imread('imssb.jpg', cv.IMREAD_GRAYSCALE)
img02=cv.imread('imssb.jpg')
```

```python
print("ancho de imagen", img01.shape[0])
print("largo de imagen", img01.shape[1])
```

```python
imshow("Imagen en niveles de gris", img01, 10)
```

```python
imshow("Imagen a color",img02)
```
