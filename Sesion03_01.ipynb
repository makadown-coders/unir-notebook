```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
plt.figure(figsize=(10,5))

# Histograma de canal azul
plt.subplot(2,2,1)
plt.plot(b_hist, color='blue')
plt.title('Histograma de Azul')
plt.xlim([0, 256])

# histograma de canal verde
plt.subplot(2,2,2)
plt.plot(b_hist, color='green')
plt.title('Histograma de Verde')
plt.xlim([0, 256])

# histograma de canal rojo
plt.subplot(2,2,3)
plt.plot(b_hist, color='red')
plt.title('Histograma de Rojo')
plt.xlim([0, 256])

# mostrar graficos
plt.tight_layout()
plt.show()
```
